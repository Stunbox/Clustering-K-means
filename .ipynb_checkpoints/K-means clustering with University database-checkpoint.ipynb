# Clustering K-means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
colleges = pd.read_csv('College_Data',index_col=0)
colleges.info() #Vemos la informacion que contiene el csv
```

```python
colleges.head() #Mostramos algunos datos
```

```python
colleges_vars = colleges.drop(['Private'],axis=1) #Removemos la primer variable para el analisis
```

```python
colleges_vars.describe() #Datos estadisticos
```

```python
colleges_norm = (colleges_vars-colleges_vars.min())/(colleges_vars.max()-colleges_vars.min()) #Normalizamos
colleges_norm
```

```python
colleges_norm.describe() #Datos ya normalizados con minimo 0 y maximo 1
```

```python
#Metodo codo de jambu para encontrar el no de clusters correspondientes
array = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i,max_iter = 300)
    kmeans.fit(colleges_norm)
    array.append(kmeans.inertia_)
```

```python
#Graficamos los datos
plt.plot(range(1,11),array)
plt.title("Codo de jambu")
plt.xlabel("Numero de clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
#Metodo K-means a la bd
clustering = KMeans(n_clusters = 5, max_iter = 300)
clustering.fit(colleges_norm)
```

```python
#Se agrega la columna al archivo original

colleges['KMeans_clusters'] = clustering.labels_
colleges.head()
```

```python
#Visualizacion de los clusters

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_colleges = pca.fit_transform(colleges_norm)

pca_colleges_df = pd.DataFrame(data = pca_colleges,columns = ['Component_1','Component_2'])
pca_colleges_df
```

```python
cluster_numbers = []
numColleges = colleges.to_numpy()

for i in range(0,777):
    cluster_numbers.append(numColleges[i][18])

cluster_numbers
```

```python
clusters_numbers_df = pd.DataFrame(data = cluster_numbers ,columns = ['KMeans_clusters'])
pca_colleges_df['KMeans_clusters'] = clusters_numbers_df['KMeans_clusters']

pca_names_colleges = pca_colleges_df
pca_names_colleges
```

```python
fig = plt.figure(figsize = (6,6))

ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Component 1', fontsize = 15)
ax.set_ylabel('Component 2', fontsize = 15)
ax.set_title('Componentes principales', fontsize = 20)

color_theme = np.array(["blue","green","orange","black","yellow"])
ax.scatter(x = pca_names_colleges.Component_1, y = pca_names_colleges.Component_2,
          c = color_theme[pca_names_colleges.KMeans_clusters],s=50)
plt.show()
```
